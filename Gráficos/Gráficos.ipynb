```python
# Vamos colocar 4 graficos
# primeiro criamos a figura e os eixos

fig, axs = plt.subplots(nrows = 2 , ncols = 2 , figsize=(5,5))
plt.tight_layout()
```

```python
# Agora vamos adicionar os gráfico aos eixos


axs[0][0].scatter(df2['alcohol'], df2['speeding'])
axs[0][1].bar(df1['sex'], df1['survived'])
axs[1][0].hist(df1['age'])
axs[1][1].boxplot(df1['age'])

fig
```

Percebemos que o matplotlib não faz automaticamente algumas das operações, como agrupamento com soma, e por algum motivo o bloxplot não saiu.

Então vamos recorrer ao seaborn e ver como são plotados os graficos em cada eixo

```python
fig2, axs2 = plt.subplots(nrows = 2 , ncols = 2 , figsize=(5,5))
plt.tight_layout() #isso é importante para não ficar muito junto


sns.scatterplot(data=df2, x='alcohol', y='speeding', ax=axs2[0][0])
sns.barplot(data= df1, x = df1['sex'], y = df1['survived'],ax=axs2[0][1])
sns.histplot(data = df1['age'], ax=axs2[1][0])
sns.boxplot(data=df1, x=df1['age'], orient='h', ax=axs2[1][1])

fig2.show()
```
